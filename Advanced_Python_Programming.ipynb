##5.4 Challenge-2 Find the Factors of a Number

If the user enters 20, output should be "The factors of 20 is 1,2,4,5,10,20"



```python
user_number = int(input("Enter a Number: "))

print(f"The factors of {user_number} are:", end=" ")

for i in range(1, user_number+1):
    if user_number % i == 0:
        print(i, end=",")
```

```python
user_input = int(input("Enter your number: "))

result = []
for i in range(1,user_input+1):
    if user_input%i==0:
        result.append(i)
print("The factors of {} are {}".format(user_input,result))
```
